```python
import pandas as pd
import numpy as np
def Bisection(Xl:float,Xu:float,Tolerance:float,maxIteration:int,F):
    assert F(Xl) * F(Xu) <0
    Error = 1.0
    Xu_Loop = Xu
    Xl_Loop = Xl
    iterations= 0
    Xr_Old = 0
    Xrs = []
    Xus=[]
    Xls=[]
    Es = []
    Fxrs =[]
    while((np.abs(Error)>=Tolerance) and (iterations)<=maxIteration):
        iterations = iterations+1
        FXl = F(Xl)
        Xr_Loop = Xu_Loop - (Xu_Loop-Xl_Loop)/2
        FXr = F(Xr_Loop)
        Decision = (FXl * FXr)
        if(Decision < 0):
            Xu_Loop = Xr_Loop
        elif(Decision > 0):
            Xl_Loop = Xr_Loop
        else:
            break
        Error = (Xr_Loop -Xr_Old)/(Xr_Loop)
        Xr_Old = Xr_Loop
        Xrs.append(Xr_Loop)
        Xus.append(Xu_Loop)
        Xls.append(Xl_Loop)
        Es.append(Error)
        Fxrs.append(FXr)
    table = pd.DataFrame({"Xr":Xrs,"Xu":Xus,"Xl":Xls,"Error":Es,"Fxr":Fxrs})
    return table
```

```python
def Function(X:float):
    return X**3 - 25
```

```python
Bisection(2,3,1e-18,100,F=Function)
```

```python
def False_Position(Xl:float,Xu:float,Tolerance:float,maxIteration:int,F):
    assert F(Xl) * F(Xu) <0
    Error = 1.0
    Xu_Loop = Xu
    Xl_Loop = Xl
    iterations= 0
    Xr_Old = 0
    Xrs = []
    Xus=[]
    Xls=[]
    Es = []
    Fxrs =[]
    while(np.abs(Error)>=Tolerance and iterations<=maxIteration):
        iterations=iterations+1
        FXl = F(Xl_Loop)
        FXu =F(Xu_Loop)
        Xr_Loop = ((Xl_Loop * FXu)-(Xu_Loop * FXl))/(FXu - FXl)
        FXr = F(Xr_Loop)
        if(np.sign(FXu)==np.sign(FXr)):
            Xu_Loop = Xr_Loop
        elif(np.sign(FXl)==np.sign(FXr)):
            Xl_Loop = Xr_Loop
        Error = (Xr_Loop - Xr_Old)/(Xr_Loop)
        Xr_Old = Xr_Loop
        Xrs.append(Xr_Loop)
        Xus.append(Xu_Loop)
        Xls.append(Xl_Loop)
        Es.append(Error)
        Fxrs.append(FXr)
    table = pd.DataFrame({"Xr":Xrs,"Xu":Xus,"Xl":Xls,"Error":Es,"Fxr":Fxrs})
    return table
```

```python
def Fixed_Position(Xo:float,Tolerance:float,maxIterations:int,Gx):
    X_Loop = Xo
    Error=1.0
    iterations=0
    X_Old=Xo
    Xs = []
    Es = []
    while(np.abs(Error)>= Tolerance and iterations<=maxIterations):
        iterations = iterations+1
        X_Old = X_Loop
        X_Loop = Gx(X_Loop)
        Error = ((X_Loop - X_Old)/X_Loop)
        Xs.append(X_Loop)
        Es.append(Error)
    table = pd.DataFrame({"Xr":Xs,"Error":Error})
    return table
        
```
